```python
# Element-wise operations
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Addition
addition = a + b
print("\nAddition:", addition)  # [5 7 9]

# Multiplication
multiplication = a * b
print("Multiplication:", multiplication)  # [ 4 10 18]

# Other operations: subtraction (-), division (/), power (**), etc.
```

```python
# Matrix operations
matrix_a = np.array([[1, 2],
                      [3, 4]])
matrix_b = np.array([[5, 6],
                    [7, 8]])

# Matrix multiplication
matrix_mult = np.dot(matrix_a, matrix_b)
print("\nMatrix Multiplication:")
print(matrix_mult)

# Output:
# Matrix Multiplication:
# [[19 22]
#  [43 50]]

# Transpose of a matrix
matrix_transpose = np.transpose(matrix_a)
print("\nTranspose of Matrix A:")
print(matrix_transpose)

# Output:
# Transpose of Matrix A:
# [[1 3]
#  [2 4]]
```

```python
# Broadcasting with arrays
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([10, 20, 30])

result_broadcast = arr1 + arr2

print("\nBroadcasting Arrays:")
print(result_broadcast)

# Output:
# Broadcasting Arrays:
# [[11 22 33]
#  [14 25 36]]
```
